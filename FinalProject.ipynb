# Finding the best community area in Chicago to live in

```python
import bs4 as bs
import urllib
import urllib.request
import pandas as pd
```

###  Web scraping and Parsing the Wikipedia and convert the data into a DataFrame

```python
Chicago_coomn_area_url = 'https://en.wikipedia.org/wiki/Community_areas_in_Chicago'
Chicago_coomn_area_urlopen = urllib.request.urlopen(Chicago_coomn_area_url)
soup = bs.BeautifulSoup(Chicago_coomn_area_urlopen,'html.parser')
#soup
```

```python
#soup.find_all('table')
```

```python
output_rows_final=[]
for commnarea in soup.find_all('table', {'class': "wikitable"}):
    output_row1=[]
    #output_rows=[]
    #print(commnarea)
    for x in commnarea.find_all('th'):
        #print(x)
        output_row1.append(x.text.strip())
    #print(output_row1)
    #output_rows.append(output_row1)
    #print(output_rows)
    for z in commnarea.find_all('tr'):
        output_row2 = []
        for y in z.find_all('td'):
            #print(y.text)
            output_row2.append(y.text.strip())
        #print("ROW 2", output_row2)
        #if "None" not in output_row2:
        if len(output_row2) > 0:
            output_row2 = [x1.replace("\n", ",") for x1 in output_row2]
            output_rows_final.append(output_row2)
    #print(output_rows)
    #output_rows_final.append(output_rows)
print("-------")
print(output_rows_final[:10])
```

```python
df=pd.DataFrame(output_rows_final, columns = output_row1, index=None)
df.columns = [ 'COMMUNITY AREA NUMBER', 'COMMUNITY AREA NAME','NEIGHBOURHOODS']

df = df.sort_values(by=['COMMUNITY AREA NUMBER']).reset_index()
df = df[['COMMUNITY AREA NAME','COMMUNITY AREA NUMBER', 'NEIGHBOURHOODS']]
df['COMMUNITY AREA NAME'] = df['COMMUNITY AREA NAME'].str.upper() 
df.head(10)
#df.shape
```

### Loading the 'Census Data' into 'DataFrame' and retain the required features

```python
df1=pd.read_csv("https://ibm.box.com/shared/static/05c3415cbfbtfnr2fx4atenb2sd361ze.csv")
df1.head(10)
```

```python
df1_v1=df1[["COMMUNITY_AREA_NAME", "PERCENT OF HOUSING CROWDED", "PERCENT HOUSEHOLDS BELOW POVERTY", "PERCENT AGED 16+ UNEMPLOYED"]]
df1_v1.columns = ['COMMUNITY AREA NAME', 'PERCENT OF HOUSING  CROWDED', 'PERCENT OF HOUSEHOLDS BELOW POVERTY', 'PERCENT AGED 16+ UNEMPLOYED']
df1_v1['COMMUNITY AREA NAME'] = df1_v1['COMMUNITY AREA NAME'].str.upper() 
df1_v1.drop([77],axis=0).head(10)
```

```python
dfn_v1=df.merge(df1_v1,how='left',on='COMMUNITY AREA NAME')
dfn_v1.head(10)
```

### Loading the 'School Data' into 'DataFrame' and retain the required features

```python
df2=pd.read_csv("https://ibm.box.com/shared/static/f9gjvj1gjmxxzycdhplzt01qtz0s7ew7.csv")
df2.head(5)
```

```python
df2_v1 = df2.groupby('COMMUNITY_AREA_NAME').agg({'Elementary, Middle, or High School': ', '.join,'COMMUNITY_AREA_NUMBER':'first'}).reset_index()
df2_v1 = df2_v1.sort_values(by=['COMMUNITY_AREA_NUMBER']).reset_index()
df2_v1 = df2_v1[['COMMUNITY_AREA_NAME','Elementary, Middle, or High School']]
df2_v1.columns = ['COMMUNITY AREA NAME', 'Elementary, Middle, or High School']
df2_v1.head(10)
```

### Extracting the 'Latitude' and 'Longitude' values of Community Areas

```python
df2_v2 = df2.groupby('COMMUNITY_AREA_NAME').agg({'Latitude':'first','Longitude':'first','COMMUNITY_AREA_NUMBER':'first'}).reset_index()
df2_v2 = df2_v2.sort_values(by=['COMMUNITY_AREA_NUMBER']).reset_index()
df2_v2 = df2_v2[['COMMUNITY_AREA_NAME','Latitude', 'Longitude']]
df2_v2.columns = ['COMMUNITY AREA NAME', 'Latitude', 'Longitude']
df2_v2.head(10)
```

```python
dfn_v2=dfn_v1.merge(df2_v1,how='left',on='COMMUNITY AREA NAME')
dfn_v2.head(10)
```

### Merging all data frames - Final Data Frame

```python
dfn_v3=dfn_v2.merge(df2_v2,how='left',on='COMMUNITY AREA NAME')
dfn_v31=dfn_v3.drop([72],axis=0)
dfn_v32=dfn_v31.drop([75],axis=0)
dfn_v33=dfn_v32.drop([17],axis=0)
#dfn_v33
dfn_v33.columns = ['Community_Area_Name','Community_Area_Number','Neighbourhoods','Housing_Crowded_Percent','Below_Poverty_Percent','Unemployment_Percent','School_Type','Latitude', 'Longitude']
dfn_v33 = dfn_v33[['Community_Area_Name', 'Neighbourhoods','Housing_Crowded_Percent', 'Below_Poverty_Percent', 'Unemployment_Percent','School_Type', 'Latitude', 'Longitude']]
dfn_v33
```

```python
#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

### Community Areas

```python
map_Chicago = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, label in zip(dfn_v33['Latitude'], dfn_v33['Longitude'], dfn_v33['Community_Area_Name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Chicago)  
    
map_Chicago
```

### K-means clustering

```python
import numpy as np
```

```python
kclusters = 4

dfn_v33_cluster = dfn_v33.drop('School_Type', 1)
dfn_v33_cluster = dfn_v33_cluster.drop('Unemployment_Percent', 1)
dfn_v33_cluster = dfn_v33_cluster.drop('Below_Poverty_Percent', 1)
dfn_v33_cluster = dfn_v33_cluster.drop('Housing_Crowded_Percent', 1)
dfn_v33_cluster = dfn_v33_cluster.drop('Neighbourhoods', 1)
dfn_v33_cluster = dfn_v33_cluster.drop('Community_Area_Name', 1)
dfn_v33_cluster.head()
```

```python
kclusters=5
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(dfn_v33_cluster)

kmeans.labels_[0:10]
```

```python
dfn_v33.insert(5,'Cluster Labels', kmeans.labels_)
dfn_v33
```

```python
dfn_v33_1 = dfn_v33.groupby('Cluster Labels').agg({'Community_Area_Name': ', '.join}).reset_index()
dfn_v33_1
```

```python
dfn_v33_2=dfn_v33.groupby('Cluster Labels').mean()
dfn_v33_2 = dfn_v33_2[['Housing_Crowded_Percent','Below_Poverty_Percent','Unemployment_Percent']]
dfn_v33_2
```

```python
dfn_v33_f=dfn_v33_2.merge(dfn_v33_1,how='left',on='Cluster Labels')
dfn_v33_f
```

```python
dfn_v33_f.corr()
```

### Clustered Community Areas

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(dfn_v33['Latitude'], dfn_v33['Longitude'], dfn_v33['Community_Area_Name'], dfn_v33['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
